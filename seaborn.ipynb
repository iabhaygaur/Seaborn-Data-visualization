part 1 :Retail Superstore Data Exploration
In this section, we ingest and explore a multi-year retail performance dataset to identify trends across geographical regions and product segments.

```python
df=pd.read_excel(r"C:\Users\edtech\Downloads\Sample - Superstore.xls")
df
```

```python
import seaborn as sns
```

```python
# Regional Sales Volume and Distribution
sns.barplot(x="Region",y="Sales",data=df, color="red")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Region",y="Sales",data=df,hue="Segment")
plt.show()
```

```python
#Product Category Counts by Region
sns.countplot(x="Category",data=df,hue="Region")
plt.show()
```

```python
df["Year"]=df["Order Date"].dt.year
```

```python
df["Year"]=df["Year"].astype(str)
```

```python
# Multi-Year Sales Trajectory by Product Category
plt.figure(figsize=(8,4))
sns.lineplot(x="Year",y="Sales",data=df,hue="Category",marker="o")
plt.show()
```

```python
# Profitability Analysis (Sales vs. Profit Margin)
plt.figure(figsize=(4,2))
sns.scatterplot(x="Sales",y="Profit",data=df,marker="o",color="black",hue="Category")
plt.show()
```

Part 2: Institutional Enrollment & Demographic Analysis
We now pivot to a second dataset containing student registration profiles to evaluate audience demographics, distribution spreads, and institutional tracking.

```python
df=pd.read_excel(r"C:\Users\edtech\Downloads\Education Data Analysis.xlsx")
df
```

```python
# Age and Gender Distribution Spreads
plt.figure(figsize=(8,6))
sns.histplot(x="Age",data=df,bins=12,kde=True,hue="Gender")
plt.show()
```

```python
# Statistical Spread and Outlier Identification
plt.figure(figsize=(6,4))
sns.boxplot(x="Age",data=df,hue="Gender")
plt.show()
```

```python
# String Vectorization & Domain Analysis
df["Domain"] = df["Email"].str.split("@").str[1]
```

```python
plt.figure(figsize=(8, 4))

domain_counts = df["Domain"].value_counts().head(5)
sns.barplot(x=domain_counts.index, y=domain_counts.values, palette="dark:blue_r")

plt.title("Top 5 Student Email Domain Distributions", fontsize=14, pad=15)
plt.xlabel("Email Domain", fontsize=12)
plt.ylabel("Number of Enrollments", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
